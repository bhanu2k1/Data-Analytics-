1. What is hypothesis testing in statistics?
-  Hypothesis method in statistics is used to evaluate claims about a population by analyzing sample data.

2. What is the null hypothesis, and how does it differ from the alternative hypothesis?
- Null Hypothesis (Ho) > Assumes there is no relationship or difference between two variables or groups. Involves = sign.
- Alternate Hypothesis (Ha) > Contradict the null hypothesis, suggesting a real difference. Involves inequality sign.

3. What is the significance level in hypothesis testing, and why is it important?
- Significance level (alpha) > probability of rejecting the null hypothesis when it is actually true. It is also known as margin of error.
- It is important in deciding the threshold whether observed results are statistically significant.

4. What does a P-value represent in hypothesis testing?
- P-value is probability value which is used to decide whether to reject a null hypothesis or not.

5.  How do you interpret the P-value in hypothesis testing?
- Small p-value > less than equal to 0.05 indicates strong evidence against null hypothesis.
- Large p-value > more than 0.05 indicates weak evidence against null hypothesis.

6. What are Type 1 and Type 2 errors in hypothesis testing?
- Type 1 error > Rejection of null hypothesis when it is actually true. ex: An innocent person is convicted.
- Type 2 error > Failure to reject the null hypothesis when it is actually false. ex: A guilty person may not be convicted.

1. Write a Python program to perform a Z-test for comparing a sample mean to a known population mean and
interpret the results.

```python
import numpy as np
from statsmodels.stats.weightstats import ztest

sample_mean = 25.5
population_mean = 25
sample_size = 50
population_std_dev = 2
alpha = 0.05

data = np.random.normal(sample_mean, population_std_dev, sample_size)
z_statistic, p_value = ztest(data, value=population_mean)
print(f"Z_statistics is {z_statistic:.4f}")
print(f"p_value is {p_value:.4f}")

if p_value < alpha:
  print("Reject the null hypothesis.")
else:
  print("Fail to reject the null hypothesis.")
```

4. Perform a two-tailed Z-test using Python and visualize the decision region on a plot.

```python
import matplotlib.pyplot as plt
from scipy.stats import norm
sample_mean = 65
population_mean = 60
population_std = 10
sample_size = 50
alpha = 0.05
standard_error = population_std/np.sqrt(sample_size)
z_statistic = (sample_mean - population_mean) / standard_error
print(f"Z_stat is {z_statistic:.3f}")
critical_z_lower = norm.ppf(alpha/2)
critical_z_upper = norm.ppf(1 - alpha/2)
print(f"Critical values are {critical_z_lower:.3f} and {critical_z_upper:.3f}")
if abs(z_statistic) > critical_z_upper:
  print("Reject the null hypothesis (H0)")
else:
  print("Failt to reject the null hypothesis (H0)")
x = np.linspace(-4,4,500)
pdf = norm.pdf(x)
plt.figure(figsize=(10,6))
plt.plot(x, pdf, color='blue', label='Standard Normal Distribution')
plt.fill_between(x, 0, pdf, where=(x < critical_z_lower) | (x > critical_z_upper),
                 color='red', label='Rejection Region')
plt.axvline(z_statistic, color='green', linestyle='--', label='Z-statistics')
plt.axvline(critical_z_lower, color='red', linestyle=':', label='Lower')
plt.axvline(critical_z_upper, color='red', linestyle=':', label='Upper')
plt.title('Two Tail Z-test Decision Region Visualization')
plt.xlabel('Z-score')
plt.ylabel('Probability Density')
plt.legend()
plt.grid(True)
plt.show()
```

5. Create a Python function that calculates and visualizes Type 1 and Type 2 errors during hypothesis testing.

```python
mu_0 = 100
sigma = 10
alpha = 0.05
mu_a = 105
x = np.linspace(mu_0-3*sigma, mu_a+3*sigma, 500)
pdf_null = stats.norm.pdf(x, loc=mu_0, scale=sigma)
pdf_alt = stats.norm.pdf(x, loc=mu_a, scale=sigma)
critical_value = stats.norm.ppf(1-alpha, loc=mu_0, scale=sigma)
plt.figure(figsize=(10,6))
plt.plot(x, pdf_null, label='Null Hypothesis Testing')
plt.plot(x, pdf_alt, label='Alternative Hypothesis Testing')
x_reject_type1 = x[x > critical_value]
plt.fill_between(x_reject_type1, 0, stats.norm.pdf(x_reject_type1,
              loc=mu_0, scale=sigma),color='red', label='Type I error')
x_accept_H0 = x[x < critical_value]
plt.fill_between(x_accept_H0, 0, stats.norm.pdf(x_accept_H0,
             loc=mu_a, scale=sigma), color='blue', label='Type II error')
plt.axvline(critical_value, color='black', linestyle='--',
            label='Critical Value')
plt.title('Type I and Type II Error Visualization')
plt.xlabel('Sample Mean')
plt.ylabel('Probability Density')
plt.grid(True)
plt.legend()
plt.show()
```

7. Perform a paired sample T-test using Python and visualize the comparison results.

```python
before = np.array([75, 80, 78, 82, 79, 76, 81, 77, 80, 83])
after = np.array([70, 75, 76, 78, 74, 72, 79, 75, 77, 79])

t_statistic, p_value = stats.ttest_rel(before, after)

print(f"T-statistic: {t_statistic:.3f}")
print(f"P-value: {p_value:.3f}")

# Interpret the results
alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis")
else:
    print("Fail to reject the null hypothesis")

plt.figure(figsize=(8, 6))

for i in range(len(before)):
    plt.plot([1, 2], [before[i], after[i]], marker='o', color='gray',
             linestyle='-', alpha=0.6)
plt.scatter(1, np.mean(before), color='blue', s=100,
            label=f'Mean Before: {np.mean(before):.2f}')
plt.scatter(2, np.mean(after), color='red', s=100,
            label=f'Mean After: {np.mean(after):.2f}')

plt.xticks([1, 2], ['Before', 'After'])
plt.ylabel('Measurement Value')
plt.title('Paired Sample Comparison (Before vs. After)')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.legend()
plt.tight_layout()
plt.show()
```

9. Write a Python function to calculate the confidence interval for a sample mean and explain its significance.

```python
def confidence_interval(data, confidence_level):
    mean = np.mean(data)
    std_error = stats.sem(data)
    if len(data) < 30:
      interval = stats.t.interval(confidence_level, len(data)-1,
                                  loc=mean, scale=std_error)
    else:
      interval = stats.norm.interval(confidence_level,
                                     loc=mean, scale=std_error)
    return interval

data = [10, 12, 15, 18, 20, 22, 25, 28, 30, 32]
confidence_level = 0.95
interval = confidence_interval(data, confidence_level)
print(f"Confidence Interval: {interval}")
```

10. Write a Python program to calculate the margin of error for a given confidence level using sample data.

```python
import numpy as np
from scipy import stats as s

def margin_of_error(data, confidence_level=0.95):
    sample_mean = np.mean(data)
    sample_std = np.std(data, ddof=1)
    sample_size = len(data)
    t_critical = s.t.ppf((1 + confidence_level) / 2, df=sample_size - 1)
    standard_error = sample_std / np.sqrt(sample_size)
    margin_of_error = t_critical * standard_error
    return margin_of_error

data = [165, 170, 172, 168, 175, 169, 171, 173, 167, 170]
margin = margin_of_error(data)
print(f"Margin of Error: {margin:.2f}")
```
