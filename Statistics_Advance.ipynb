THEORY QUESTIONS

1. What is a random variable in probability theory?
- Random variable is set of possible values for a random experiment.

2. What are the types of random variables?
- Discrete Random Variable > Possible values are whole number.
eg. Tossing of a coin, Rolling of a dice.
- Continous Random Variable > Take any value within a specified range or interval.
eg. Height of students, Temprature in a room.

3. What is the difference between discrete and continuous distributions?
- Discrete Distribution > Deals with countable and seperated values. eg. 0,1,2,3.
- Continous Distribution > Deals with any variables that can take any value within a given range. eg. 180.1, 182.4, 183.6

4. What are probability distribution functions (PDF)?
- Probability Distribution Functions describes the probability of a random variable taking on different values.

5. How do cumulative distribution functions (CDF) differ from probability distribution functions (PDF)?
- CDF > Gives probability that a random variable is less than or equal to a specific value.
- PDF >  Gives probability of a random variable taking a specific value.

6. What is a discrete uniform distribution?
- A type of probability distribution where a finite number of outcomes are equally likely to occur.
- Mean, median and mode is same.
- Symmetrical distribution with skewness equal to zero.

7. What are the key properties of a Bernoulli distribution?
- Binary outcomes i.e. 1 as success and 0 as failure.
- Each trial should be independent.
- Constant probabilty of success (p) for each trial.

8. What is the binomial distribution, and how is it used in probability?
- Binomial distribution is 'n' bernoulli trials.
- It is used in probability to find out responses in a binary outcome when experiment is done 'n' number of times.

9. What is the Poisson distribution and where is it applied?
- Poisson distribution describes no. of events that occur within a fixed interval of time and space given a known avg. rate of occurenece.
- Applied for number of people visiting any place.

10. What is a continuous uniform distribution?
- Probabilty distribution where a continuous random variable is equally likely to take any value within a specific range (a,b).

11. What are the characteristics of a normal distribution?
- Symmetrical about mean i.e. skewness = 0
- mean = median = mode
- Follows empirical rule of normal distribution. (std = Standard Deviation)
-- 68% of data lies within 1 std
-- 95& of data lies within 2 std
-- 99.7% of data lies within 3 std

12. What is the standard normal distribution, and why is it important?
- Standard Normal Distribution is a continous probability distribution.
- A special case of normal distribution where mean = 0 and std = 1.
- It is important to compare data from different normal distributions and enables probabilt calculations using Z scores.

13. What is the Central Limit Theorem (CLT), and why is it critical in statistics?
- CLT states that distribution of sample means will be approximate normal distribution.
- CLT is fundamental in machine learning, particularly for validating models.
- Critical for statistical methods like hypothesis testing.

14. How does the Central Limit Theorem relate to the normal distribution?
- Distribution of sample means will approach a normal distribution, regardless of shape of population distribution as the sample size gets larger.

15. What is the application of Z statistics in hypothesis testing?
- To determine if there is a significant difference between a sample mean and population mean.
- To find difference between the means of two samples when certain conditions are met.

16. How do you calculate a Z-score, and what does it represent?
- Z score is known as standard score which is calculated by subtracting the mean from a data point and then dividing it by standard deviation.

17. What are point estimates and interval estimates in statistics?
- Point estimates > A single value calculated from sample data that is used to estimate a population parameter.
- Interval estimates > Provides a range of values within which the population parameter is likely to fall.

18. What is the significance of confidence intervals in statistical analysis?
- Confidence intervals are crucial because they provide a range of plausible values for an unknown population parameter, indicating the uncertainty associated with sample estimates.

19. What is the relationship between a Z-score and a confidence interval?
- The Z-score, which represents the number of standard deviations a data point is from the mean, is used to calculate the margin of error when constructing a confidence interval.

20. How are Z-scores used to compare different distributions?
- Z-scores are used to compare data from different distributions by standardizing the values.

21. What are the assumptions for applying the Central Limit Theorem?
- Random Sampling > Sample must be drawn randomly from the population.
- Independent Samples > Observation within the sample must be independent of each other.
- Sufficient large samples > Sample size must be greater than equal to 30.
- Sampling without replacements

22. What is the concept of expected value in a probability distribution?
- The expected value in a probability distribution represents the long-run average outcome of a random variable.

23. How does a probability distribution relate to the expected outcome of a random variable?
- Probability distribution describes the likelihood of all possible outcomes of a random variable.
- Expected outcome is weighted average of these outcomes.

9. Write a Python function to calculate Z-scores from a dataset and plot them.

```python
def calculate_z_score(data):
  mean = np.mean(data)
  std_dev = np.std(data)
  z_score = (data - mean) / std_dev
  return z_score
data = np.array([1,2,3,4,5])
z_score = calculate_z_score(data)
plt.plot(z_score)
plt.title('Z-Score Plot')
plt.xlabel('Data Points')
plt.ylabel('Z-Scores')
plt.grid()
plt.show()
```

13. Generate random variables and calculate their corresponding probabilities using the binomial distribution.

```python
import math
def binomial_probability(n,k,p):
  combinations = math.comb(n,k)
  probability = combinations * (p**k) * ((1-p)**(n-k))
  return probability
n = 5 # no of trials
k = 3 # no of successes
p = 0.7 # probabilty of success in a single trial
binomial_probability(n,k,p)
```

14. Write a Python program to calculate the Z-score for a given data point and compare it to a standard normal
distribution.

```python
import numpy as np
from scipy.stats import norm
def calculate_z_score(data_point, mean, std_dev):
  z_score = (data_point - mean) / std_dev
  return z_score
def compare(z_score):
  cumulative_probability = norm.cdf(z_score)
  extreme_probability = 2*(1-norm.cdf(abs(z_score)))
  return cumulative_probability, extreme_probability
data_set = np.array([10, 12, 15, 18, 20, 22, 25, 28, 30, 32])
mean = np.mean(data_set)
std_dev = np.std(data_set)
data_point = 20
compare(calculate_z_score(data_point, mean, std_dev))
```

17. Generate data from a normal distribution, then calculate and interpret the confidence interval for its mean.

```python
import scipy.stats
sample_size = 50
mean = 10
std_dev = 2
data = np.random.normal(mean, std_dev, sample_size)
sample_mean = np.mean(data)
sample_std_dev = np.std(data, ddof=1)
confidence_level = 0.95
z_score = scipy.stats.norm.pdf((1+ confidence_level)/2)
margin_of_error = z_score*(sample_std_dev/np.sqrt(sample_size))
confidence_interval = (sample_mean - margin_of_error, sample_mean + margin_of_error)
print(confidence_interval)
```

23. Implement Z-score normalization on a dataset and visualize its transformation.

```python
original_data = np.random.normal(loc=10, scale=5, size=1000)
normalized_data = (original_data - np.mean(original_data)) / np.std(original_data)
plt.figure(figsize=(10,8))
plt.subplot(2,1,1)
plt.hist(original_data, bins=30, label='Original Data')
plt.title("Original Data Distribution")
plt.xlabel("Value")
plt.ylabel("Frequency")

plt.subplot(2,1,2)
plt.hist(normalized_data, bins=30, label='Normalized Data')
plt.title("Normalized Data Distribution")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()
```
